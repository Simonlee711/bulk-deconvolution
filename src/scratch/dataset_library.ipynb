```python
# use glob to get all the csv files 
# in the folder

path = '../../../../../../cbbi-data00/projects/Methods/bulk-deconvolution/signatures/'
csv_files = glob.glob(os.path.join(path, "*.compartment_level.csv"))

l = []
cell_type = []

df = pd.read_csv(csv_files[0],index_col = 0)
for columns in df.columns:
    l.append(df[columns].to_list())
    length = df[columns].count()
    for i in range(length):
        cell_type.append(columns)

path = '../../../../../../cbbi-data00/projects/Methods/bulk-deconvolution/'
csv_files = glob.glob(os.path.join(path, "*.celltype_level.csv"))

# loop over the list of csv files
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f, index_col = 0)
    index = df.index
    col = df.columns

    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    for columns in df.columns:
        l.append(df[columns].to_list())
        length = df[columns].count()
        for i in range(length):
            cell_type.append(columns)


    display(df)

final = [item for sublist in l for item in sublist]
final = [x for x in final if str(x) != 'nan']

print(len(final), '==', len(cell_type))
```

# gene length 

```python
# use glob to get all the csv files 
# in the folder
path = '../../../../../../cbbi-data00/projects/Methods/bulk-deconvolution/'
csv_files = glob.glob(os.path.join(path, "*.celltype_level.csv"))
  
final_df = pd.DataFrame(columns = ['Cell_type', 'length'])
# loop over the list of csv files
counter = 0
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f, index_col = 0)
      
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    display(df)
    
    for col in df.columns:
        print("[%s]" % ', '.join(map(str, df[col].unique())))
        final_df.loc[counter] = [col, len(df[col].dropna())] 
        counter = counter + 1


display(final_df) 
```

# true percentages

```python
# takes differential results and returns the top X genes from each compartment
df = pd.read_csv("../../../../../../cbbi-data00/projects/Methods/bulk-deconvolution/markers.compartment_level.csv")
df = df.drop("Unnamed: 0",axis=1)
df

for end in [20,40,60,80]:
    l = []
    temp = df.iloc[:end, :]
    for col in temp.columns:
        l.append(temp[col].unique().tolist())
    
    genes = [item for sublist in l for item in sublist]

    df2 = pd.read_csv("../../../../../../cbbi-data00/projects/Methods/bulk-deconvolution/marker_gene_expression.by_compartments.csv",index_col = 0)
    df2 = df2.drop("Unknown")
    df2 = df2[df2.columns.intersection(genes)]
    df2.to_csv("../../data/genesize/compsig_"+str(end)+".csv")
```
